```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url='https://cafely.com/blogs/research/which-country-consumes-the-most-coffee'
data=requests.get(url)
soup = BeautifulSoup(data.text, 'html')
print(soup)
```

```python
table=soup.find_all('table')[1]
print(table)
```

```python
coffee_titles = table.find_all('th')
coffee_table_titles = [title.text.strip() for title in coffee_titles]

print(coffee_table_titles)
```

```python
coffee_table_titles=coffee_table_titles[1:]
```

```python
print(coffee_table_titles)
```

```python
df=pd.DataFrame(columns=coffee_table_titles)
df
```

```python
column_data=table.find_all('tr')
print(column_data)
```

```python
for row in column_data:
    row_data=row.find_all('td')
    individual_row_data=[data.text.strip() for data in row_data]
    individual_row_data=individual_row_data[1:]
    length=len(df)
    print(individual_row_data)
    if(len(individual_row_data)!=0):
        df.loc[length]=individual_row_data
df
```
